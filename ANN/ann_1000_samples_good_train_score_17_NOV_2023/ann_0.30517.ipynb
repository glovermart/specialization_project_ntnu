```python
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
import numpy as np
import sys
import matplotlib.pyplot as plt 
np.set_printoptions(threshold=sys.maxsize)
```

```python
csv1 = pd.read_csv('sequence_execution_time.csv')
#csv1.head()
```

```python
csv1.tail()
```

```python
sns.boxplot (data = csv1.execution_time)
```

```python
csv1 = csv1[csv1.execution_time < 100000]
```

```python
sns.boxplot (data = csv1.execution_time)
```

```python
#data = csv1.drop(['sequence_identifier','quality'], axis = 1)
data = csv1
```

```python
data1 = data.drop(csv1.index[1000:]) 
#data.head()
```

```python
data1.tail()
```

```python
reads = data1.drop(['sequence_identifier','quality'], axis =1)
reads
```

```python
#Normalize data
reads['execution_time'] = np.log10(reads.execution_time)
reads
```

```python
def Kmers_funct(seq, size=9):
   return [seq[x:x+size].upper() for x in range(len(seq) - size + 1)]
reads['words'] = reads.apply(lambda x: Kmers_funct(x['read_sequence']), axis=1)
reads = reads.drop('read_sequence', axis=1)
```

```python
human_texts = list(reads['words'])
for item in range(len(human_texts)):
   human_texts[item] = ' '.join(human_texts[item])
#separate labels
Y = reads.iloc[:, 0].values # y_human for human_dna
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range = (4,4)) # initially (4,4)
X = cv.fit_transform(human_texts)
```

```python
X = X.todense()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,Y, test_size = 0.2, random_state = 1) 
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
#y_test
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Create a model with multiple hidden layers. Note that input and output has
# dimension one
#M = 16 #64
model = keras.Sequential([
    keras.layers.Dense(4, activation=tf.nn.relu, input_dim = X_train.shape[1]),
    keras.layers.Dense(6, activation=tf.nn.relu), #128
    keras.layers.Dense(8, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.relu),
    keras.layers.Dense(12, activation=tf.nn.relu),
    keras.layers.Dense(14, activation=tf.nn.relu),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(18, activation=tf.nn.relu),
    keras.layers.Dense(20, activation=tf.nn.relu),
    keras.layers.Dense(22, activation=tf.nn.relu),
    keras.layers.Dense(24, activation=tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(96, activation=tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(126, activation=tf.nn.relu),
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(512, activation=tf.nn.relu),
    #keras.layers.Dropout(0.5),
    #keras.layers.Dense(128, activation=tf.nn.relu),
   # keras.layers.Dense(64, activation=tf.nn.relu), #256
#   keras.layers.Dense(256, activation=tf.nn.relu),
    #keras.layers.Dropout(0.5),
    #keras.layers.Dense(128, activation=tf.nn.relu),#512
   #keras.layers.Dense(M, activation=tf.nn.relu),
   #keras.layers.Dense(M, activation=tf.nn.relu),
   # keras.layers.Dropout(0.5),
    keras.layers.Dense(1) #, activation=tf.nn.relu # , activation= 'linear'
])
model.summary()

# Train the model
model.compile(loss='mean_squared_error',
              optimizer="adam",
              metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=50)
```

```python
#history = model.fit(X_train, y_train, epochs=300, batch_size=128, verbose=True, validation_data = (X_test,y_test))
history = model.fit(X_train, y_train, epochs=1000, batch_size= 8, verbose=True, validation_split = 0.2,  callbacks = [callback])
```

```python
# Visualize the training results
plt.figure(figsize=(10,8))
plt.plot(history.history['loss'],label='loss')
#plt.plot(history.history['accuracy'],label='acc')
plt.plot(history.history['val_loss'],label='val_loss')
#plt.plot(history.history['val_accuracy'],label='val_acc')
plt.grid()
plt.legend()
plt.show()
```
